## 3. Bivariate Analysis
Customer profile(characteristics of a customer) of the different products 
### 3.1 Categorical vs Categorical 
Compare 'Product' and other categorical datasets(Gender,Marital Status) by using grouped bar chart analysis

```python
#Product VS Gender
plt.subplot(121) #(1x2, 1st)
data.Gender.value_counts(normalize=True).plot(kind = 'bar',figsize=(15,10)); #count gender values and plot as a bar
plt.tight_layout(pad=10); #graph layout
plt.xlabel('Gender',labelpad=15); #x-axis name
plt.xticks(rotation='horizontal'); #set xtick label horizontal
plt.ylabel('# of products consumed', labelpad=15); #y-axis name

#Product VS Marital Status
plt.subplot(122) #(1x2, 2nd)
data.MaritalStatus.value_counts(normalize=True).plot(kind = 'bar',figsize=(15,10)); #count maritalstatus values and plot as a bar
plt.tight_layout(pad=5); #graph layout
plt.xlabel('Marital Status',labelpad=15); #x-axis name
plt.xticks(rotation='horizontal'); #set xtick label horizontal
plt.ylabel('# of products consumed', labelpad=15); #y-axis name
```

### 3.2 Categorical vs Numerical 
Compare Product and other numerical dataset(Age, Education,Usage, Income, Miles) to see the data distributions and densitys.

```python
# set graphs figure size 
data.plot(figsize=(25,25));

#Product vs Age
plt.subplot(231) #(2x3, 1st)
sns.swarmplot(data['Age'], data['Product'], color='Black', edgecolor='white', size=8); 
plt.tight_layout(pad=5);

sns.boxplot(x='Age',y='Product',data=data); 

#Product vs Education
plt.subplot(232) #(2x3, 2nd)
sns.swarmplot(data['Product'], data['Education'], color='Pink', edgecolor='black', size=8);
plt.tight_layout(pad=5);

sns.violinplot(x="Product", y="Education",inner=None, data=data);

#Product vs Usage
plt.subplot(233) #(2x3, 3rd)
sns.pointplot(data['Product'], data['Usage'], size=8);
plt.tight_layout(pad=5);


#Product vs Income
plt.subplot(234) #(2x3, 4th)
sns.barplot(data['Product'], data['Income']);
plt.tight_layout(pad=5);

sns.swarmplot(data['Product'], data['Income'], size=8, color='white', edgecolor='black');

#Product vs Miles
plt.subplot(235) #(2x3, 5th)
sns.swarmplot(data['Product'], data['Miles'], size=8);
plt.tight_layout(pad=5);
```

### 4.2 Heatmap
Correlations among pairs of variables

```python
data.plot(figsize=(9,9));
sns.heatmap(data.corr(), annot=True, linewidths=.5, center = 1);  
```

#### 4.2.2 Conclusions
- There are top three correlations:
   1. ' Miles',' Usage': It is very likely that if the customer is expected to run more miles, the number of times using treadmill will increase
   2. 'Miles','Fitness': It is very likely that if the customer is expected to run more miles, the customer would rate it's fitness score higher
   - Because both 'Usage' and 'Fitness' have high correlations with 'Miles', they are correlated also.If the customer is expected to run more miles, the customer would rate it's fitness score higher
   3. 'Income', 'Education': It is very likely that if the customer gets paied higher, the customer spent longer years in education
